```python
class LayerNorm(nn.Module):
    def __init__(self,input):
        super().__init__()
        self.alpha=torch.ones((input.size(-1)))
        self.beta=torch.zeros((input.size(-1)))
        self.eps=1e-12
    def forward(self,x):
        mean=torch.mean(x,dim=-1).unsqueeze(-1)
        var= ((torch.sum(((x - mean)**2),dim=-1))/x.size(-1)).unsqueeze(-1)
        output = self.alpha*((x-mean)/torch.sqrt(var+self.eps)) + self.beta
        return output
```

The journey was something like this:
Sigmoid -> Tanh -> ReLU -> GeLU -> SwiGLU
Sigmoid(x)=1/1+e^-x ( used in binary classification)
Tanh -> centered at zero, but still if a neuron falls into the staright region it's hard to recover
ReLU -> max(0,x) -> Dead neuron problems
GELU -> x * φ(x) the function is gaussian cumulative distributive function
