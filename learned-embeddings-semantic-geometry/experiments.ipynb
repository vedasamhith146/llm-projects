```python
with open("data/tiny.txt",'r') as f:
    text=f.read().lower()
from collections import Counter

def next_char_distribution(ch,text):
    counter=Counter()
    for ch1,ch2 in zip(text,text[1:]):
        if ch1==ch:
            counter[ch2]+=1
    total=sum(counter.values())
    return {k:v/total for k,v in counter.items()}
```

```python
stats_a=next_char_distribution('a',text)
stats_a_list=list(stats_a.items())
stats_a_list.sort(key=lambda stats_a_list:stats_a_list[1],reverse=True)
print(stats_a_list[:10])
```

```python
stats_i=next_char_distribution('i',text)
stats_i_list=list(stats_i.items())
stats_i_list.sort(key=lambda stats_i_list:stats_i_list[1],reverse=True)
print(stats_i_list[:10])
```

```python
from collections import Counter
import torch
import json

with open("stoi.json",'r') as f:
    stoi=json.load(f)

def probability_vector(ch,text,stoi):
    counter=Counter()
    for c1, c2 in zip(text, text[1:]):
        if c1 == ch:
            counter[c2] += 1

    total = sum(counter.values())
    vec=torch.zeros(len(stoi))

    for nxt_char, count in counter.items():
        vec[stoi[nxt_char]]= count/total
    return vec
```

```python
a_vec=probability_vector('a',text,stoi)
i_vec=probability_vector('i',text,stoi)

sim=torch.dot(a_vec,i_vec)/(torch.norm(a_vec)*torch.norm(i_vec))
print(sim.item())
```
