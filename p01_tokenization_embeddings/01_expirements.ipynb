```python
import torch
import torch.nn.functional as f
import torch.nn as nn
```

```python
one_hot_run=f.one_hot(torch.tensor(410),num_classes=1264)
one_hot_walk=f.one_hot(torch.tensor(119),num_classes=1264)
one_hot_jump=f.one_hot(torch.tensor(106),num_classes=1264)
```

```python
import numpy as np
```

```python
def cosine_similarity(a,b):
    a=np.array(a)
    b=np.array(b)
    return np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
```

```python
print("cosine similarity between run and walk using one hot vectors is",cosine_similarity(one_hot_run,one_hot_walk))
```
